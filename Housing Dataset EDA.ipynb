Fixing Column Names and Changing Column Data Types

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'[^\w]', '', regex=True)
```

```python
df.rename(columns={'bedrooms': 'num_bedrooms', 'bathrooms': 'num_bathrooms', 'yearbuilt': 'year_built', 'areasqft': 'area_sqft', 'lotsize': 'lot_size', 'daysonmarket': 'days_on_market', 'propertytype': 'property_type'}, inplace=True)
```

```python
df['num_bathrooms'] = df['num_bathrooms'].str.replace(r'\D+', '', regex=True)
df['num_bedrooms'] = df['num_bedrooms'].str.replace(r'\D+', '', regex=True)
df['area_sqft'] = df['area_sqft'].str.replace(r'\D+', '', regex=True)
df['lot_size'] = df['lot_size'].str.replace(r'\D+', '', regex=True)
```

```python
df.dtypes.to_dict
```

```python
df.drop(['mls_id', 'listing_agent', 'listing_url'], axis=1, inplace=True)
```

```python
df.dtypes
```

```python
int_col = ['num_bedrooms', 'area_sqft', 'lot_size', 'price']
for col in int_col:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(r'[^\d]', '', regex=True)
        .replace('', np.nan)
        .astype('Int64')
    )
```

```python
df['num_bathrooms'] = df['num_bathrooms'].astype(str).str.replace(r'[^0-9.]', '', regex=True).replace('', np.nan).astype(float)
```

```python
df['address'] = df['address'].astype('string')
df['zipcode'] = df['zipcode'].astype('string')
```

```python
cat_cols = ['city', 'state', 'property_type', 'status']
for col in cat_cols:
    df[col] = df[col].astype('category')
```
